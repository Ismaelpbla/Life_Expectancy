```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as ss
```

## DATA 

```python
df = pd.read_csv('all_data.csv')
head = df.head()
print(head)

tail = df.tail(3)
print(tail)

length = len(df)
print( 'The total lenght of the Dataframe is: ',  length)

info = df.info()
print(info)

description = df.describe()
print(description)

unique_countries = df.Country.unique()

unique_year = df.Year.unique()

print(unique_countries)
print(unique_year)

cv = ss.variation(df['Life expectancy at birth (years)'])
cv2 = ss.variation(df['GDP'])

print(cv)
print(cv2)
```

## DATA CLEANING CHECKING

```python
# Missing Data Method
cols = df.columns
colours = ['#000099', '#ffff00']
sns.heatmap(df[cols].isnull(), cmap=sns.color_palette(colours))
```

```python
duplicated_data = df.duplicated()

print(duplicated_data.value_counts())
```

```python
df = df.rename(columns={'Life expectancy at birth (years)': 'Life_exp'})
df['Country'] = df['Country'].replace({'United States of America':'USA'})
print(df.head())
print(df['Country'].value_counts())
```

## DATA ANALYSIS

```python
Chile = df[df['Country'] == 'Chile']
Germany = df[df['Country'] == 'Germany']
USA = df[df['Country'] == 'USA']
Mexico = df[df['Country'] == 'Mexico']
China = df[df['Country'] == 'China']
Zimbabwe = df[df['Country'] == 'Zimbabwe']

plt.figure(figsize=(25,25))

plt.subplot(3,2,1)
sns.lineplot (x= 'Year', y= 'Life_exp', data=Chile)
plt.title('Chile')
plt.subplot(3,2,2)
sns.lineplot (x= 'Year', y= 'Life_exp', data=Germany)
plt.title('Germany')
plt.subplot(3,2,3)
sns.lineplot (x= 'Year', y= 'Life_exp', data=USA)
plt.title('USA')
plt.subplot(3,2,4)
sns.lineplot (x= 'Year', y= 'Life_exp', data=Mexico)
plt.title('Mexico')
plt.subplot(3,2,5)
sns.lineplot (x= 'Year', y= 'Life_exp', data=China)
plt.title('China')
plt.subplot(3,2,6)
sns.lineplot (x= 'Year', y= 'Life_exp', data=Zimbabwe)
plt.title('Zimbabwe')
```

```python
plt.figure(figsize= (25, 25))
sns.lineplot (x= 'Year', y= 'Life_exp', data=Chile)
sns.lineplot (x= 'Year', y= 'Life_exp', data=Germany)
sns.lineplot (x= 'Year', y= 'Life_exp', data=China)
sns.lineplot (x= 'Year', y= 'Life_exp', data=Mexico)
sns.lineplot (x= 'Year', y= 'Life_exp', data=Zimbabwe)
sns.lineplot (x= 'Year', y= 'Life_exp', data=USA)
plt.legend()
```
